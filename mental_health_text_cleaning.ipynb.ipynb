```python
import pandas as pd
import json
```

```python
try:
    with open('intents.json', 'r') as f:
        raw_data = json.load(f)
except FileNotFoundError:
    raw_data = {"intents": [
        {"tag": "anxious", "patterns": ["I feel overwhelmed", "My heart is racing"], "responses": ["Take a deep breath."]},
        {"tag": "sad", "patterns": ["I want to cry", "Nothing goes right"], "responses": ["I am here for you."]}
    ]}

    rows = []
for intent in raw_data['intents']:
    tag = intent['tag']
    # Loop through each pattern a human typed out
    for pattern in intent['patterns']:
        rows.append({'User_Emotion_Tag': tag, 'User_Text': pattern})

df = pd.DataFrame(rows)
df['User_Text'] = df['User_Text'].str.lower().str.strip()  # Clean formatting strings
df['Text_Word_Count'] = df['User_Text'].apply(lambda x: len(str(x).split())) # Feature Extraction

df.to_csv('cleaned_mental_health_metrics.csv', index=False)

print("Step 1 Complete! Flattened table structure preview:")
print(df.head())
```
